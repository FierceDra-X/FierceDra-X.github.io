```python
def exchange_rates(currency, format='full'):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = req.get(url).json()['Valute']    
    data = response[currency]     
    if format == 'full':    
        return data      
    elif format == 'value':    
        return data['Value']
```

```python
exchange_rates('UAH', 'value')
```

```python
def currency_name(id):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = req.get(url).json()['Valute']
    currency = response.keys()
    for currency in currency:
        if response[currency]['ID'] == id:
            return response[currency]['Name']
```
